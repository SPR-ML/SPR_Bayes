```python
import pandas as pd
import xlrd
stu_inf = pd.read_csv('./student.csv',encoding = 'gb18030')
stu_inf.head()
```

```python
stu_inf.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
stu_inf['school'] = le.fit_transform(stu_inf['school'])
stu_inf['age'] = le.fit_transform(stu_inf['age'])
stu_inf['sex'] = le.fit_transform(stu_inf['sex'])
stu_inf['address'] = le.fit_transform(stu_inf['address'])
stu_inf['famsize'] = le.fit_transform(stu_inf['famsize'])
stu_inf['Pstatus'] = le.fit_transform(stu_inf['Pstatus'])
stu_inf['Mjob'] = le.fit_transform(stu_inf['Mjob'])
stu_inf['Fjob'] = le.fit_transform(stu_inf['Fjob'])
stu_inf['reason'] = le.fit_transform(stu_inf['reason'])
stu_inf['guardian'] = le.fit_transform(stu_inf['guardian'])
stu_inf['schoolsup'] = le.fit_transform(stu_inf['schoolsup'])
stu_inf['famsup'] = le.fit_transform(stu_inf['famsup'])
stu_inf['paid'] = le.fit_transform(stu_inf['paid'])
stu_inf['activities'] = le.fit_transform(stu_inf['activities'])
stu_inf['nursery'] = le.fit_transform(stu_inf['nursery'])
stu_inf['higher'] = le.fit_transform(stu_inf['higher'])
stu_inf['internet'] = le.fit_transform(stu_inf['internet'])
stu_inf['romantic'] = le.fit_transform(stu_inf['romantic'])

stu_inf.head(20)
```

```python
import re
import nltk

# label = list(stu_inf['G']//10)
label = list(stu_inf['G']//5)
label[2]
```

```python
x=['G1','G2','G3','G']
featuresets = stu_inf.drop(stu_inf[x],axis=1)

featuresets.head()
```

```python
from sklearn.model_selection import train_test_split
text_train,text_test,labels_train,labels_test=train_test_split(featuresets,label,test_size=0.25,random_state=123)
text_train.head()
# labels_train
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
clf1=GaussianNB()
clf2=MultinomialNB()
clf1.fit(text_train,labels_train)
clf2.fit(text_train,labels_train)
#pre=clf1.predict(text_test)
print("GaussianNB :{}".format(clf1.score(text_test,labels_test)))
print("MultinomialNB :{}".format(clf2.score(text_test,labels_test)))
```
